```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv("penguins.csv")
```

```python
dataset.head()
```

```python
print(dataset.describe())
```

```python
numeric_features = dataset.select_dtypes(include=[np.number]).columns.tolist()

plt.figure(figsize=(12, 6))
dataset[numeric_features].boxplot()
plt.title('Wykres pudekowy dla cech numerycznych')
plt.xticks(rotation=45)
plt.show()
```

```python
# spraawdzam nulle
missing_values = dataset.isnull().sum()
print(missing_values)
```

```python
# Wykres rozproszenia dla dwch cech dugoci dzioba i zakrzywienia
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=dataset)
plt.title('Wykres rozproszenia midzy dugoci dzioba a gbokoci dzioba')
plt.show()
```

```python
# Histogramy dla cech numerycznych dlalepszego widzenia
dataset.hist(figsize=(10, 8))
plt.suptitle('Histogramy cech numerycznych')
plt.show()
```

```python
#oraz wykres pudekowy
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='flipper_length_mm', data=dataset)
plt.title('Wykres pudekowy dla dugoci skrzyda wedug gatunku pingwinw')
plt.show()
```

```python
# Wiersze zawierajce wartoci null
rows_with_null = dataset[dataset.isnull().any(axis=1)]
print(rows_with_null)
```

```python
# Wykres pudekowy na mass
plt.figure(figsize=(8, 6))
sns.boxplot(x=dataset['body_mass_g'])
plt.title('Wykres pudekowy dla masy ciaa')
plt.xlabel('Body Mass (g)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=dataset['bill_length_mm'])
plt.title('dlugo dzioba')
plt.xlabel('dlugo dzioba')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=dataset['bill_depth_mm'])
plt.title('skrzywienie/wgbienie dzioba')
plt.xlabel('skrzywienie/wgbienie dzioba')
plt.show()
```
